### Requirements

```python
import os
from glob import glob
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
from PIL import Image
```

```python
images_original_path = "/home/Ravikumar/freelance/computer vision/music_genre/dataset/images_original"
```

### Categories of Musics

```python
classes_names = os.listdir(images_original_path)
num_classes = len(classes_names)
print(f"Total Number of Categories : {num_classes}\n")
print(classes_names)
```

```python
rows, columns = 2, 5

fig, axes = plt.subplots(rows, columns, figsize=(10, 5))


for idx, cls in enumerate(classes_names):
    path = f"/home/Ravikumar/freelance/computer vision/music_genre/dataset/images_original/{cls}"
    image_path = os.path.join(path, os.listdir(path)[0])
    image = np.array(Image.open(image_path))
    row = idx // columns
    col = idx % columns
    axes[row, col].imshow(image, interpolation='bicubic')
    axes[row, col].set_title(cls)

plt.show()
plt.savefig("/home/Ravikumar/freelance/computer vision/music_genre/results/sample.png")
```

```python
counts = list()
for cls in classes_names:
    path = os.path.join(f"/home/Ravikumar/freelance/computer vision/music_genre/dataset/images_original/{cls}")
    c= len(os.listdir(path))
    counts.append(c)
    # print(f"Number of {cls} Samples: {len(os.listdir(path))}")
fig = px.bar(counts,color = classes_names)
fig.show()
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load audio file using Librosa
audio_file = '/home/Ravikumar/freelance/computer vision/music_genre/dataset/genres_original/blues/blues.00000.wav'
y, sr = librosa.load(audio_file)

# Calculate spectrogram using Librosa
spec = librosa.stft(y)
spec_db = librosa.amplitude_to_db(abs(spec))

# Plot spectrogram using Matplotlib
plt.figure(figsize=(10, 4))
librosa.display.specshow(spec_db, x_axis='time', y_axis='log')
plt.colorbar()
plt.title('Spectrogram')
plt.tight_layout()
plt.show()
```

```python
y, s = librosa.load(audio_file)


audio, _ = librosa.effects.trim(y)

print('Audio File:', audio, '\n')
print('Audio File shape:', np.shape(audio))

```

```python


plt.figure(figsize = (16, 6))
librosa.display.waveshow(y = audio, sr = s, color = "#00008B");
plt.title("Example Sound Waves on Blues", fontsize = 23);

```
